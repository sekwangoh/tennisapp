## 넥젠이 등장한 2011년부터  2021년까지 데이터 가져오기

```python
# 2011~2021년 데이터 가져오기!
path = './next_generation'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')]
```

```python
%cd next_generation
```

```python
# 데이터 가져옴
df_tennis = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(i) ## 각 csv 파일
    df_tennis = pd.concat([df_tennis, data]) ## 하나의 데이터 프레임을 붙이기
```

### 데이터 정보가 담긴 사전을 확인하여 필요없는 데이터 제거하기!

```python
#### 제거할 특성
- tourney_id : 경기의 고유 id 값
- match_num : 그 대회에 몇 번 째 경기인지 알려주는 값   
- winner_id : 이긴 선수의 고유 id 값
- winner_seed : 이긴 선수의 그 대회 시드순위(시드는)
- winner_entry : 이긴 선수가 어떤 자격으로 출전하게 되었는지 (- 'WC' = wild card, 'Q' = qualifier, 'LL' = lucky loser, 'PR' = protected ranking, 'ITF' = ITF entry, and there are a few others that are occasionally used.)
- winner_ioc : 이긴 선수의 국적
- loser_id
- loser_seed
- loser_entry
- loser_ioc


### 남겨놓을 특성
- tourney_name : 투어 경기의 이름
- surface : 코트의 특성
- tourney_level : 대회의 크기와 상금 규모를 확인할 수 있는 지표(그랜드 슬램, 마스터즈, ATP 투어대회 등등) (For men: 'G' = Grand Slams, 'M' = Masters 1000s, 'A' = other tour-level events, 'C' = Challengers, 'S' = Satellites/ITFs, 'F' = Tour finals and other season-ending events, and 'D' = Davis Cup )
- tourney_date : 투어가 이루어진 년/월/일 (eight digits, YYYYMMDD, usually the Monday of the tournament week.)
- winner_name : 이긴 선수의 이름
- winner_hand : 이긴 선수가 오른손 잡이인지 왼손잡이 인지(- R = right, L = left, U = unknown. For ambidextrous players, this is their serving hand.)
- winner_ht : 이긴선수의 키

- score : 세트 결과 (ex. 6-4,6-2,6-3)
- best_of : 최대 세트 수 (그랜드 슬램 5세트 경기, 마스터즈 등 기타 대회 3세트 경기)
- round : 그 경기가 몇 라운드 인지(ex그랜드 슬램 기준, R128-1라운드, R64-2라운드, R32-3라운드, R16-4라운드, QF-준준결승, SF-준결승, F-결승)
- minutes : 경기 진행 시간   
- w_ace : 이긴 선수의 위너수 (winner's number of aces)
- w_df : 이긴 선수의 더블폴트 수 (winner's number of doubles faults)
- w_svpt : 이긴 선수의 서브 포인트 수 (winner's number of serve points
- w_1stIn : 이긴 선수의 첫 서브 성공률(winner's number of first serves made)
- w_1stWon : 이긴 선수의 첫 서브 성공 시 이긴 게임 수 (winner's number of first-serve points won)
- w_2ndWon : 이긴 선수의 세컨 서브 성공시 이긴 게임 수 (winner's number of second-serve points won)
- w_SvGms : 서브 게임수 (winner's number of serve games)
- w_bpSaved : 브레이크 포인트에서 막아낸 수 (winner's number of break points saved)
- w_bpFaced : 브레이크 포인트에 직면한 수 (winner's number of break points faced)
- winner_rank : 이긴 선수의 세계 랭킹 (winner's ATP or WTA rank, as of the tourney_date, or the most recent ranking date before the tourney_date
- winner_rank_points : 이긴 선수의 세계 랭킹 포인트 (number of ranking points, where available)                               
                               

### 진선수들    
- loser_name
- loser_hand
- loser_ht                               
- l_ace
- l_df                     
- l_svpt
- l_1stIn
- l_1stWon
- l_2ndWon
- l_SvGms
- l_bpSaved
- l_bpFaced
- loser_rank
- loser_rank_points                               
                               
    
### 고민되는 특성
- draw_size : 경기의 크기를 이야기 해준다. 그랜드 슬램의 경우 128강부터 시작이고, 투어급 대회는 크기에 따라 다르다.
- winner_age : 투어 대회 당시 이긴 선수의 나이
- loser_age
```
